# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

# **Bar** Chart

```python
numerical_cols = df.select_dtypes(include='number').columns


df[numerical_cols].hist(bins=30, figsize=(15, 10), edgecolor='black')
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.show()
```

# Interpretation

Presence of Skewness(Negative and Positive respectively)

Presence of outliers


# **Box Plot**

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.suptitle('Boxplots of Numerical Features', fontsize=16)
plt.subplots_adjust(top=0.92)
plt.show()
```

# Iterpretation

Presence of outliers in all the cols

# ●	Bar Chart Analysis for Categorical Column:

```python
categorical_cols = categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5 * len(categorical_cols)))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(len(categorical_cols), 1, i)
    sns.countplot(data=df, x=col, palette='pastel', order=df[col].value_counts().index)
    plt.title(f'Count of Categories in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.suptitle('Bar Chart Analysis of Categorical Columns', fontsize=16)
plt.subplots_adjust(top=0.95)
plt.show()
```

# Insights

1)Most sales occurred on Thursday

2)The SKU column shows a uniform distribution, with each SKU appearing the same number of times.
3) All sales recorded in City C
4)Each BU appears the same number of times.
5)jeera dominates the sales, having the highest count among all brands. The remaining brands are unevenly distributed but all appear less frequently than Jeera.
6)All models are evenly distributed in the dataset.

# **Standardization of Numerical Variables**

Standardization is a feature scaling technique that transforms numerical variables to have a mean of 0 and a standard deviation of 1 using the Z-score

```python
Numericals_cols= df.select_dtypes(include='number').columns
```

```python
df[numerical_cols].describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df_standardized[numerical_cols].describe()
```

# Before and after comparisons of the data distributions.

```python

for col in numerical_cols:
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, color='navy')
    plt.title(f'Before Standardization: {col}')

    plt.subplot(1, 2, 2)
    sns.histplot(df_standardized[col], kde=True, color='darkred')
    plt.title(f'After Standardization: {col}')

    plt.tight_layout()
    plt.show()
```
