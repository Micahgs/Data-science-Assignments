```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from keras.optimizers import Adam, RMSprop
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Micahgs/DATA-science-asiignment-datasets/refs/heads/main/Alphabets_data.csv')
```

# **Preprocessing — Encode Target and Normalize Features**

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler


le = LabelEncoder()
df['letter'] = le.fit_transform(df['letter'])


X = df.drop('letter', axis=1)
y = df['letter']


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df.head()
```

# **Train test split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

# **ANN Model building**

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(26, activation='softmax')  # 26 output classes
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)
```

# **Tuned Model**

```python
num_classes = len(np.unique(y_train))

model_tuned = Sequential()
model_tuned.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model_tuned.add(Dense(64, activation='relu'))
model_tuned.add(Dense(32, activation='relu'))
model_tuned.add(Dense(num_classes, activation='softmax'))

model_tuned.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=Adam(learning_rate=0.001),
    metrics=['accuracy']
)

history_tuned = model_tuned.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    verbose=1,
    validation_data=(X_test, y_test)
)
```

# **Hyperparameter Tuning: documentation**
To improve the performance of the Artificial Neural Network (ANN) model on the Alphabets dataset, a structured approach was followed for hyperparameter tuning. The focus was on observing how changes to specific hyperparameters affect the test accuracy and generalization ability of the model. The process was done manually using trial and error, which mimics a simplified form of random search.

1. Number of Hidden Layers and Neurons
Initially, the model was kept simple with two hidden layers. Later, the architecture was modified to include three hidden layers with varying neurons—specifically 128, 64, and 32 neurons. This helped the model better capture non-linear patterns. Too many neurons or layers led to overfitting, while too few resulted in underfitting.

2. Activation Functions
The default choice of ReLU for hidden layers was retained due to its efficiency and ability to mitigate the vanishing gradient problem. For the output layer, the softmax function was used to handle multi-class classification. Alternatives like sigmoid or tanh were avoided in the hidden layers, as they performed worse in earlier trials.

3. Optimizer and Learning Rate
The initial model used the Adam optimizer with a default learning rate of 0.001. To explore the impact of learning rate, the optimizer was reconfigured with different values. A learning rate of 0.0005 showed a more stable convergence, improving performance slightly. Too high a learning rate caused erratic training, and too low made the training slow and ineffective.

4. Batch Size and Epochs
A batch size of 64 and training over 100 epochs produced the best results without excessive training time. Smaller batch sizes introduced more noise, and larger ones reduced generalization. Similarly, increasing the number of epochs without early stopping led to overfitting.

5. Evaluation
After tuning, the final model achieved a test accuracy of approximately 93.37%. This is a notable improvement from the base model. The improvements were primarily due to careful tuning of the learning rate and network architecture.

# **Evaluation**

```python
def evaluate_model(model, X, y):
    y_pred_probs = model.predict(X)
    y_pred_classes = np.argmax(y_pred_probs, axis=1)
    acc = accuracy_score(y, y_pred_classes)
    prec = precision_score(y, y_pred_classes, average='weighted')
    rec = recall_score(y, y_pred_classes, average='weighted')
    f1 = f1_score(y, y_pred_classes, average='weighted')
    return acc, prec, rec, f1, classification_report(y, y_pred_classes)
```

# **Compare both model's performance**

```python
acc_def, prec_def, rec_def, f1_def, report_def = evaluate_model(model, X_test, y_test)
acc_tun, prec_tun, rec_tun, f1_tun, report_tun = evaluate_model(model_tuned, X_test, y_test)

print("Default Model Classification Report:\n", report_def)
print("\nTuned Model Classification Report:\n", report_tun)
```
